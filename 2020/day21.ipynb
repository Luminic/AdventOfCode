```python
with open("day21_input.txt", "r") as f:
    prog_inp = f.readlines()

prog_inp = [line.rstrip() for line in prog_inp]

for i in range(len(prog_inp)-1,-1,-1):
    if prog_inp[i] == '': del prog_inp[i]
    else: break
print(prog_inp[:5])
```

```python
# Tuples in the form (ingredient list, allergen list)
parsed_inp = []
all_ingredients = set()
all_allergens = set()
for line in prog_inp:
    ingredients, allergens = line.split(" (")
    ingredients = ingredients.split(' ')
    allergens = allergens[9:-1].split(", ")
    parsed_inp.append((ingredients, allergens))
    for i in ingredients:
        all_ingredients.add(i)
    for a in allergens:
        all_allergens.add(a)

print(all_ingredients)
print(all_allergens)
parsed_inp
```

```python
class Entree:
    def __init__(self, allergen_list, inp_index):
        self.allergen_list = allergen_list
        self.inp_index = inp_index

    def __repr__(self):
        return str(self.allergen_list) + ' ' + str(self.inp_index)
```

```python
possibilities = dict()
# confirmed = dict()
confirmed_ingredients = dict()
for i, line in enumerate(parsed_inp):
    for allergen in line[1]:
        if allergen in possibilities:
            possibilities[allergen] &= set(line[0])
        else:
            possibilities[allergen] = set(line[0])

print(possibilities)

while 1:
    change = False
    uncertain_allergens = list(possibilities.keys())
    for i, allergen in enumerate(uncertain_allergens):
        if len(possibilities[allergen]) == 1:
            # confirmed[allergen] = next(iter(possibilities[allergen]))
            confirmed_ingredients[next(iter(possibilities[allergen]))] = allergen
            del possibilities[allergen]
            change = True
        else:
            for possible_containing_ingredient in list(possibilities[allergen]):
                if possible_containing_ingredient in confirmed_ingredients:
                    possibilities[allergen].remove(possible_containing_ingredient)
                    change = True
    if not change:
        break

print("\nAfter:")
print("possibilities:", possibilities)
print("confirmed_ingredients", confirmed_ingredients)
```

```python
allergen_free = all_ingredients - set(confirmed_ingredients.keys())
print("allergen_free:", allergen_free)

occurences_of_allergen_free = 0
for ingredients, allergens in parsed_inp:
    for ingredient in ingredients:
        if ingredient in allergen_free:
            occurences_of_allergen_free += 1
print("occurences of allergen free ingredients:", occurences_of_allergen_free)
```
