```python
# Boilerplate: loading input

prog_inp = ""
with open("day6_input.txt", "r") as f:
    prog_inp = f.readlines()[:-1]

# prog_inp = [line.rstrip() for line in prog_inp]
# print(prog_inp[:5])
# print(prog_inp[-5:])
print(prog_inp)
```

```python
# Parsing input

parsed_inp = "".join(prog_inp).split("\n\n")
for (i, line) in enumerate(parsed_inp):
    parsed_inp[i] = line.strip().split('\n')

print(parsed_inp)
```

```python
from collections import defaultdict
```

```python
counts_one_plus = []
counts_all = []
for group in parsed_inp:
    counts_one_plus.append(0)
    counts_all.append(0)
    yesses = defaultdict(lambda:0)
    for form in group:
        for char in form:
            yesses[char] += 1
    for k in yesses.keys():
        if yesses[k] == len(group):
            counts_all[-1] += 1
        counts_one_plus[-1] += 1
print(sum(counts_one_plus), sum(counts_all))
```
